```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df["Gender"].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=5, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['EstimatedSalary'], bins=5, kde=True)
plt.title('Estimated Salary Distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Gender', data=df)
plt.title('Age vs. Estimated Salary')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Purchased', data=df)
plt.title('Gender vs. Purchased')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Purchased')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Purchased', y='Age', data=df)
plt.title('Age vs. Purchased')
plt.xlabel('Purchased')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Purchased', y='EstimatedSalary', data=df)
plt.title('Estimated Salary vs. Purchased')
plt.xlabel('Purchased')
plt.ylabel('Estimated Salary')
plt.show()
```
